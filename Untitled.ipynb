```python
import numpy as np
import matplotlib.pyplot as plt

from skimage.color import rgb2gray
from skimage.transform import resize
from scipy.ndimage.interpolation import shift
import scipy
from numpy.fft import *
```

```python
psf = rgb2gray(plt.imread("./PSF_crop.png"))
psf = psf / np.max(psf)
psf = resize(psf, (1096, 1096))
```

```python
weight = plt.imread("./weight.png")
```

```python
b = fft2(psf) @ fft2(weight)
```

```python
b = fftshift(np.real(ifft2(b)))
```

```python
b = (b + abs(np.min(b)))
```

```python
b  = 1/np.max(b) * b
```
